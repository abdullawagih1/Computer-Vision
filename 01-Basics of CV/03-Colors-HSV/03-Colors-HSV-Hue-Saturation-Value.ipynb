### Color space is HSV 
- H: 0 - 180
- S: 0 - 255 
- V: 0 - 255

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('img.jpg')

hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

plt.figure(figsize=(12, 8))

plt.subplot(1, 4, 1)
plt.imshow(hsv_img)
plt.axis('off')
plt.title('HSV')

plt.subplot(1, 4, 2)
plt.imshow(cv2.cvtColor(hsv_img[:, :, 0], cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('HUE')

plt.subplot(1, 4, 3)
plt.imshow(cv2.cvtColor(hsv_img[:, :, 1], cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('SATURATION')

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(hsv_img[:, :, 2], cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('VALUE')

plt.tight_layout()
plt.show()
```
