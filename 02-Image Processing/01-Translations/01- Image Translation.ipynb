## Let's code it..

## Translation

```python
img = plt.imread('img.jpg')
```

```python
height, width = img.shape[:2]
quarter_height, quarter_width = height/4, width/4
```

```python
# translation matrix
T = np.float32([[1.2, 0, quarter_width-50], 
                [0, 0.9, quarter_height-50]])
```

```python
# use cv2.warpAffine function
img_translation = cv2.warpAffine(img, T, (width, height))
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Normal')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_translation)
plt.axis('off')
plt.title('Translation')

plt.tight_layout()
plt.show()
```

## Shearing

```python
# searing matrix
Shear_mat = np.float32([[1, 0.2, 0], 
                [0.3, 1, 0]])
```

```python
Shearing_Image = cv2.warpAffine(img, Shear_mat, (width, height))
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.axis('off')
plt.title('Normal')

plt.subplot(1, 2, 2)
plt.imshow(Shearing_Image)
plt.title('Shearing')
plt.axis('off')

plt.tight_layout()
plt.show()
```
