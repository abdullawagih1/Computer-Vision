# Bitwise Operations and Masking

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### To demonstrate these operations let's create some simple images

```python
plt.figure(figsize=(10, 10))

plt.subplot(1, 2, 1)
# Rectangle
image_rec = np.zeros((512, 512), np.uint8)
cv2.rectangle(image_rec, (110, 50), (400, 300), 255, -1)
plt.imshow(cv2.cvtColor(image_rec, cv2.COLOR_BGR2RGB))
plt.title('Rectangle')
plt.axis('off')

plt.subplot(1, 2, 2)
# Ellipse
image_eli = np.zeros((512, 512), np.uint8)
cv2.ellipse(image_eli, (170, 170), (150, 120), 45, 0, 180, 100, -1)
plt.imshow(cv2.cvtColor(image_eli, cv2.COLOR_BGR2RGB))
plt.title('Ellipse')
plt.axis('off')

plt.show()
```

### NOTE: Two images will be displayed according to order you need

```python
plt.figure(figsize=(10, 10))

plt.subplot(1, 2, 1)
image = np.zeros((512, 512), np.uint8)
cv2.ellipse(image, (170, 170), (150, 120), 45, 0, 180, 100, -1)
cv2.rectangle(image, (110, 50), (400, 300), 255, -1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
image2 = np.zeros((512, 512), np.uint8)
cv2.rectangle(image2, (110, 50), (400, 300), 255, -1)
cv2.ellipse(image2, (170, 170), (150, 120), 45, 0, 180, 100, -1)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```

```python
# Shows only where they intersect
plt.figure(figsize=(8, 8))

And = cv2.bitwise_and(image_rec, image_eli)
plt.imshow(cv2.cvtColor(And, cv2.COLOR_BGR2RGB))
plt.title('And Operator')
plt.show()
```

```python
# Shows only where they union
plt.figure(figsize=(8, 8))

Or = cv2.bitwise_or(image_rec, image_eli)
plt.imshow(cv2.cvtColor(Or, cv2.COLOR_BGR2RGB))
plt.title('Or Operator')
plt.show()
```

```python
plt.figure(figsize=(8, 8))

Xor = cv2.bitwise_xor(image_rec, image_eli)
plt.imshow(cv2.cvtColor(Xor, cv2.COLOR_BGR2RGB))
plt.title('Xor Operator')
plt.show()
```

```python
plt.figure(figsize=(8, 8))

nt = cv2.bitwise_not(image_rec)
plt.imshow(cv2.cvtColor(nt, cv2.COLOR_BGR2RGB))
plt.title('Not Rectangle')
plt.show()
```
