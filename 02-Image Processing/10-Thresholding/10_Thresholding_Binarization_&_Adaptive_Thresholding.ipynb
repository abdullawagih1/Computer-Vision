```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(10, 10))

thr = cv2.imread('threshold.png')
plt.imshow(cv2.cvtColor(thr, cv2.COLOR_BGR2RGB))
plt.title('Threshold Types')
plt.axis('off')

plt.show()
```

```python
image = cv2.imread('image.png')
```

```python
plt.figure(figsize=(10, 10))

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(15, 10))

# Normal Image
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')

# Binary Threshold
plt.subplot(2, 3, 2)
ret1,thresh1 = cv2.threshold(image, 170, 255, cv2.THRESH_BINARY)
plt.imshow(cv2.cvtColor(thresh1, cv2.COLOR_BGR2RGB))
plt.title('Binary')
plt.axis('off')

# Inverted Binary Threshold
plt.subplot(2, 3, 3)
ret2,thresh2 = cv2.threshold(image, 170, 255, cv2.THRESH_BINARY_INV)
plt.imshow(cv2.cvtColor(thresh2, cv2.COLOR_BGR2RGB))
plt.title('Inverted Binary')
plt.axis('off')

# Trunkated Threshold
plt.subplot(2, 3, 4)
ret3, thresh3 = cv2.threshold(image, 170, 255, cv2.THRESH_TRUNC)
plt.imshow(cv2.cvtColor(thresh3, cv2.COLOR_BGR2RGB))
plt.title('Trunkated')
plt.axis('off')

# To Zero Threshold
plt.subplot(2, 3, 5)
ret4, thresh4 = cv2.threshold(image, 170, 225, cv2.THRESH_TOZERO)
plt.imshow(cv2.cvtColor(thresh4, cv2.COLOR_BGR2RGB))
plt.title('TOZERO')
plt.axis('off')

# Inverted To Zero Threshold
plt.subplot(2, 3, 6)
ret5, thresh5 = cv2.threshold(image, 170, 225, cv2.THRESH_TOZERO_INV)
plt.imshow(cv2.cvtColor(thresh5, cv2.COLOR_BGR2RGB))
plt.title('Inverted TOZERO')
plt.axis('off')

plt.show()
```

### Adaptive Thresholding

```python
# We will read image as gray
image = cv2.imread('image.png', 0)
```

```python
plt.figure(figsize=(10, 10))

# It's good practice to blur images as it removes noise
Gauss = cv2.GaussianBlur(image, (3, 3), 0)

# Using adaptiveThreshold
thresh = cv2.adaptiveThreshold(Gauss, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 3, 5) 

# Normal Image
plt.subplot(2, 1, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')

plt.subplot(2, 1, 2)
plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_BGR2RGB))
plt.title('Adaptive Mean Thresholding')
plt.axis('off')

plt.show()
```
