# Dilation, Erosion, Opening and Closing 
 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# We will read image as a binary for acurate results
image = cv2.imread('image.jpg', 0)
```

```python
plt.figure(figsize=(8, 8))

plt.imshow(cv2.cvtColor(image, cv2.COLOR_GRAY2RGB))
plt.title('Original Image')
plt.axis('off')

plt.show()
```

## Dilation

- Dilation increases the white region in the image.  

```python
# Create kernal
kernel = np.ones((3,3), np.uint8)
```

```python
# Let's define our kernel size
kernel = np.ones((5,5), np.uint8)
```

```python
plt.figure(figsize=(8, 8))

dilation = cv2.dilate(image, kernel, iterations = 1)
plt.imshow(cv2.cvtColor(dilation, cv2.COLOR_GRAY2RGB))
plt.title('Dilation')
plt.axis('off')

plt.show()
```

## Erosion

- Erosion is useful for removing small white noises but it shrinks the object.  

```python
plt.figure(figsize=(8, 8))

erosion = cv2.erode(image, kernel, iterations = 1)
plt.imshow(cv2.cvtColor(erosion, cv2.COLOR_GRAY2RGB))
plt.title('Erosion')
plt.axis('off')

plt.show()
```

###  Erosion is followed by Dilation as since noise is gone, they won't come back, but our object area increases. 

```python
plt.figure(figsize=(8, 8))

erosion = cv2.erode(image, kernel, iterations = 1)
dilation = cv2.dilate(erosion, kernel, iterations = 1)

plt.imshow(cv2.cvtColor(dilation, cv2.COLOR_GRAY2RGB))
plt.title('Enhanced Image')
plt.axis('off')

plt.show()
```

## Opening 
- Good for removing noise

```python
plt.figure(figsize=(8, 8))

opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
plt.imshow(cv2.cvtColor(opening, cv2.COLOR_GRAY2RGB))
plt.title('Opening')
plt.axis('off')

plt.show()
```

## Closing
- Good for removing noise

```python
plt.figure(figsize=(8, 8))

closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)

plt.imshow(cv2.cvtColor(closing, cv2.COLOR_BGR2RGB))
plt.title('Closing')
plt.axis('off')

plt.show()
```
