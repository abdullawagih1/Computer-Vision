```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('image.jpg', 0)

rows, columns = img.shape
new_img = []
```

##  Bit Plane Slicing

```python
for r in range(rows):
    for c in range(columns):
        new_img.append(np.binary_repr(img[r, c], width= 8))

bit_1 = (np.array([int(i[0]) for i in new_img], dtype= np.uint8) * 128).reshape(rows, columns)
bit_2 = (np.array([int(i[1]) for i in new_img], dtype= np.uint8) * 64).reshape(rows, columns)
bit_3 = (np.array([int(i[2]) for i in new_img], dtype= np.uint8) * 32).reshape(rows, columns)
bit_4 = (np.array([int(i[3]) for i in new_img], dtype= np.uint8) * 16).reshape(rows, columns)
bit_5 = (np.array([int(i[4]) for i in new_img], dtype= np.uint8) * 8).reshape(rows, columns)
bit_6 = (np.array([int(i[5]) for i in new_img], dtype= np.uint8) * 4).reshape(rows, columns)
bit_7 = (np.array([int(i[6]) for i in new_img], dtype= np.uint8) * 2).reshape(rows, columns)
bit_8 = (np.array([int(i[7]) for i in new_img], dtype= np.uint8) * 1).reshape(rows, columns)
```

```python
_img1 = cv2.hconcat([bit_1, bit_2, bit_3, bit_4])
_img2 = cv2.hconcat([bit_5, bit_6, bit_7, bit_8])
f_img = cv2.vconcat([_img1, _img2])

plt.figure(figsize=(12, 8))

plt.imshow(cv2.cvtColor(f_img, cv2.COLOR_BGR2RGB))
plt.title('Enhanced Images')
plt.axis('off')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
mul = bit_1 + bit_3 + bit_5 + bit_7
plt.imshow(cv2.cvtColor(mul, cv2.COLOR_BGR2RGB))
plt.title('Enhanced Image')
plt.axis('off')

plt.show()
```
