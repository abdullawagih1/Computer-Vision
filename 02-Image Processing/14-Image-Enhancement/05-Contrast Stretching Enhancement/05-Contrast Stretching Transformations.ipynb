```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('image.jpg', 0)
rows, columns = img.shape
new_img = np.array(img)
```

## Contrast Stretching

```python
# (r1, s1)    (r2, s2)    (r3, s3)    (r4, s4)
# (0, 0)      (70, 100)   (120, 200)  (255, 255)
# Constrast_Stretch => if img[r, c] >= r1 and img[r, c] <= r2: r-r1 * s2-s1 / r2-r1 + s1
```

```python
for r in range(rows):
    for c in range(columns):
        if (img[r, c] >= 0 and img[r, c] <= 70):    # r1 : r2
            new_img[r, c] = round((((img[r, c] - 0) * (100 - 0)) / (70 - 0)) + 0)    # r-r1 * s2-s1 / r2-r1 + s1

        elif (img[r, c] > 70 and img[r, c] <= 100):    # r1 : r2
            new_img[r, c] = round((((img[r, c] - 70) * (200 - 100)) / (120 - 70)) + 100)    # r-r1 * s2-s1 / r2-r1 + s1

        elif (img[r, c] > 120 and img[r, c] <= 255):   # r1 : r2
            new_img[r, c] = round((((img[r, c] - 70) * (200 - 100)) / (120 - 70)) + 100)    # r-r1 * s2-s1 / r2-r1 + s1
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB))
plt.title('Enhanced Image')
plt.axis('off')

plt.show()
```
