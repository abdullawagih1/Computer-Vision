```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('image.jpg')
```

## Power-Law (Gamma) Transformations

```python
gamma_list = [0.1, 0.5, 1.2, 2.2]
c = 255 / np.log(1 + np.max(image))

plt.figure(figsize=(12, 8))

plt.subplot(1, 4, 1)
gamma_corrected = np.array(c * (image / 255) ** gamma_list[0], dtype = 'uint8')
plt.imshow(cv2.cvtColor(gamma_corrected, cv2.COLOR_BGR2RGB))
plt.title(f'Transformation with gamma = {gamma_list[0]} ')
plt.axis('off')   

plt.subplot(1, 4, 2)
gamma_corrected = np.array(c * (image / 255) ** gamma_list[1], dtype = 'uint8')
plt.imshow(cv2.cvtColor(gamma_corrected, cv2.COLOR_BGR2RGB))
plt.title(f'Transformation with gamma = {gamma_list[1]} ')
plt.axis('off')   

plt.subplot(1, 4, 3)
gamma_corrected = np.array(c * (image / 255) ** gamma_list[2], dtype = 'uint8')
plt.imshow(cv2.cvtColor(gamma_corrected, cv2.COLOR_BGR2RGB))
plt.title(f'Transformation with gamma = {gamma_list[2]} ')
plt.axis('off')   

plt.subplot(1, 4, 4)
gamma_corrected = np.array(c * (image / 255) ** gamma_list[3], dtype = 'uint8')
plt.imshow(cv2.cvtColor(gamma_corrected, cv2.COLOR_BGR2RGB))
plt.title(f'Transformation with gamma = {gamma_list[3]} ')
plt.axis('off')   

plt.tight_layout()
plt.show()
```
