# Convolutions and Blurring

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(7, 10))

ker = plt.imread('Kernals.jpg', 1)
plt.imshow(cv2.cvtColor(ker, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Kernels Matrices')
plt.show()
```

### Convolutions Bluring (Averaging)
- Averaging done by convolving the image with a normalized box filter. 
- This takes the pixels under the box and replaces the central element. 
- Box size needs to odd and positive 

```python
img = cv2.imread('image.jpg', 1)
```

```python
# Box Blur Kernels
kernel_3x3 = np.ones((3, 3), np.float32) / 9
kernel_7x7 = np.ones((7, 7), np.float32) / 49
```

```python
plt.figure(figsize=(10, 10))

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Normal')
plt.show()
```

```python
# Kernel with 3x3
plt.figure(figsize=(10, 10))

blur3_img = cv2.filter2D(img, -1, kernel_3x3)
plt.imshow(cv2.cvtColor(blur3_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Kernal 3X3')
plt.show()
```

```python
# Kernel with 7x7
plt.figure(figsize=(10, 10))

blur7_img = cv2.filter2D(img, -1, kernel_7x7)
plt.imshow(cv2.cvtColor(blur7_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Kernal 7X7')
plt.show()
```

#### NOTE: We can use *cv2.blur()* insted of *cv2.filter2D* in averaging blur [cv2.blur(image, (3,3))]

### Gaussian Blur

```python
# Gaussian with 7x7
plt.figure(figsize=(10, 10))

Gaussian = cv2.GaussianBlur(img, (7,7), 0)
plt.imshow(cv2.cvtColor(Gaussian, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Gaussian Blur')
plt.show()
```

### Median Blur

```python
# Median with 7x7
# Takes median of all the pixels under kernel area and central element is replaced with this median value.

plt.figure(figsize=(10, 10))

Median = cv2.medianBlur(img, 5)
plt.imshow(cv2.cvtColor(Median, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Median Blur')
plt.show()
```

### Bilateral Filter

```python
# Gaussian with 7x7
# Bilateral is very effective in noise removal while keeping edges sharp
plt.figure(figsize=(10, 10))

Bilateral = cv2.bilateralFilter(img, 9, 75, 75)
plt.imshow(cv2.cvtColor(Bilateral, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Bilateral Blur')
plt.show()
```

## Image De-noising - Non-Local Means Denoising

```python
# Parameters, after None are - the filter strength 'h' (5-10 is a good range)
# Next is hForColorComponents, set as same value as h again
plt.figure(figsize=(10, 10))

dst = cv2.fastNlMeansDenoisingColored(img, None, 6, 6, 7, 21)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Fast Means Denoising')
plt.show()
```

**There are 4 variations of Non-Local Means Denoising:**

- cv2.fastNlMeansDenoising() - works with a single grayscale images
- cv2.fastNlMeansDenoisingColored() - works with a color image.
- cv2.fastNlMeansDenoisingMulti() - works with image sequence captured in short period of time (grayscale images)
- cv2.fastNlMeansDenoisingColoredMulti() - same as above, but for color images.
